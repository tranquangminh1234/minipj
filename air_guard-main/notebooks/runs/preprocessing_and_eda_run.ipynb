```python
# Cell 1: Imports
import sys
import os
import pandas as pd
# Thêm đường dẫn để import được thư viện trong src
sys.path.append(os.path.abspath(os.path.join('..')))
from src.classification_library import load_data, clean_data

# Cell 2: Parameters (Papermill sẽ ghi đè cell này)
# tags=["parameters"]
RAW_ZIP_PATH = '../data/raw/PRSA2017_Data_20130301-20170228.zip'
OUTPUT_CLEANED_PATH = '../data/processed/cleaned.parquet'
LAG_HOURS = [1, 3, 24]
USE_UCIMLREPO = False

# Cell 3: Main Execution
print("--- BẮT ĐẦU BƯỚC 1: PREPROCESSING ---")

# 1. Load Data
print(f"Đang tải dữ liệu từ: {RAW_ZIP_PATH}")
df = load_data(RAW_ZIP_PATH)
print(f"Dữ liệu thô: {df.shape}")

# 2. Clean Data
print("Đang làm sạch dữ liệu...")
df_clean = clean_data(df)
print(f"Dữ liệu sau làm sạch: {df_clean.shape}")

# 3. Lưu kết quả
os.makedirs(os.path.dirname(OUTPUT_CLEANED_PATH), exist_ok=True)
df_clean.to_parquet(OUTPUT_CLEANED_PATH)
print(f"Đã lưu file cleaned tại: {OUTPUT_CLEANED_PATH}")
```
